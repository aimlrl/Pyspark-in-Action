```python
collection = [1,"two",3.0,("four",4),{"five":5}]
```

```python
sc = spark.sparkContext
```

```python
collection_rdd = sc.parallelize(collection)
```

```python
print(collection_rdd)
```

```python
def add_one(value):
  return value + 1
```

```python
collection_rdd = collection_rdd.map(add_one)
```

```python
print(collection_rdd.collect())
```
